```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
renewables = pd.read_csv("API_EG.FEC.RNEW.ZS_DS2_en_csv_v2_13732.csv", skiprows=4)
co2 = pd.read_csv("API_EN.GHG.CO2.PC.CE.AR5_DS2_en_csv_v2_14473.csv", skiprows=4)
```

```python
# Renewable Energy Consumption (% of total final energy consumption)
renewables.head()
```

```python
# CO₂ Emissions per Capita (excluding land use)
co2.head()
```

```python
# Filter the datasets to keep only the regions of interest
regions_of_interest = [
    "European Union",
    "North America",
    "Middle East & North Africa",
    "Latin America & Caribbean",
    "Sub-Saharan Africa",
    "East Asia & Pacific"
]

renewables_regions = renewables[renewables["Country Name"].isin(regions_of_interest)].reset_index(drop=True)
co2_regions = co2[co2["Country Name"].isin(regions_of_interest)].reset_index(drop=True)
```

```python
renewables_regions.head()
```

```python
co2_regions.head()
```

```python
# Remove unwanted columns
columns_to_drop = ["Country Code", "Indicator Name", "Indicator Code", "Unnamed: 68"]
renewables_regions_cleaned = renewables_regions.drop(columns=[col for col in columns_to_drop if col in renewables_regions.columns])
co2_regions_cleaned = co2_regions.drop(columns=[col for col in columns_to_drop if col in co2_regions.columns])
```

```python
renewables_regions_cleaned.head()
```

```python
co2_regions_cleaned.head()
```

```python
# Identify years columns that have no missing values in both datasets
valid_years = [
    year for year in renewables_regions_cleaned.columns[1:]
    if renewables_regions_cleaned[year].notna().all() and co2_regions_cleaned[year].notna().all()
]

# Keep only those valid years (plus the "Country Name" column)
renewables_final = renewables_regions_cleaned[["Country Name"] + valid_years]
co2_final = co2_regions_cleaned[["Country Name"] + valid_years]
```

```python
renewables_final.head()
```

```python
co2_final.head()
```

```python
# Melt into long format
renewables_long = renewables_final.melt(id_vars="Country Name", var_name="Year", value_name="Renewables (%)")
co2_long = co2_final.melt(id_vars="Country Name", var_name="Year", value_name="CO2 per capita")
```

```python
# Merge the two datasets
merged_df = pd.merge(renewables_long, co2_long, on=["Country Name", "Year"])
```

```python
merged_df.head(10)
```

```python
merged_df["Year"] = merged_df["Year"].astype(int)

# Set up the figure
plt.figure(figsize=(14, 6))

# Plot 1: Renewable Energy (%)
plt.subplot(1, 2, 1)
sns.lineplot(data=merged_df, x="Year", y="Renewables (%)", hue="Country Name")
plt.title("Renewable Energy Over Time")
plt.xlabel("Year")
plt.ylabel("Renewables (%)")
plt.legend(title="Region", fontsize='small', title_fontsize='medium')

# Plot 2: CO2 Emissions per Capita
plt.subplot(1, 2, 2)
sns.lineplot(data=merged_df, x="Year", y="CO2 per capita", hue="Country Name")
plt.title("CO₂ Emissions per Capita Over Time")
plt.xlabel("Year")
plt.ylabel("CO₂ per Capita (tons)")
plt.legend(title="Region", fontsize='small', title_fontsize='medium')

plt.tight_layout()
plt.show()
```

```python
# Filter base year values (1990)
base_year = 1990
base_values = merged_df[merged_df["Year"] == base_year][["Country Name", "Renewables (%)", "CO2 per capita"]]
base_values = base_values.set_index("Country Name")

# Create columns for change from base year
def compute_change_from_base(row, column_name):
    base = base_values.loc[row["Country Name"], column_name]
    return ((row[column_name] - base) / base) * 100 if pd.notnull(base) else None

merged_df["Renewables Change from 1990 (%)"] = merged_df.apply(lambda row: compute_change_from_base(row, "Renewables (%)"), axis=1)
merged_df["CO2 Change from 1990 (%)"] = merged_df.apply(lambda row: compute_change_from_base(row, "CO2 per capita"), axis=1)

# Plot the trends
plt.figure(figsize=(14, 6))

# Plot 1: Renewables % change from 1990
plt.subplot(1, 2, 1)
sns.lineplot(data=merged_df, x="Year", y="Renewables Change from 1990 (%)", hue="Country Name")
plt.title("Renewables: % Change from 1990")
plt.axhline(0, color='gray', linestyle='--')
plt.ylabel("% Change from 1990")
plt.xlabel("Year")

# Plot 2: CO2 % change from 1990
plt.subplot(1, 2, 2)
sns.lineplot(data=merged_df, x="Year", y="CO2 Change from 1990 (%)", hue="Country Name")
plt.title("CO₂ per Capita: % Change from 1990")
plt.axhline(0, color='gray', linestyle='--')
plt.ylabel("% Change from 1990")
plt.xlabel("Year")

plt.tight_layout()
plt.show()
```

```python
# scatter plot: Renewables (%) vs CO2 per capita
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=merged_df,
    x="Renewables (%)",
    y="CO2 per capita",
    hue="Country Name"
)

plt.title("Renewables (%) vs CO₂ Emissions per Capita")
plt.xlabel("Renewables (%)")
plt.ylabel("CO₂ per Capita")
plt.legend(title="Region", fontsize='small', title_fontsize='medium')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Correlation between Renewables (%) and CO2 per capita overall
overall_corr = merged_df[["Renewables (%)", "CO2 per capita"]].corr().iloc[0, 1]

# Correlation per region
region_corrs = (
    merged_df.groupby("Country Name")[["Renewables (%)", "CO2 per capita"]]
    .corr()
    .iloc[0::2, 1]
    .reset_index()[["Country Name", "CO2 per capita"]]
    .rename(columns={"Country Name": "Region", "CO2 per capita": "Correlation"})
)
```

```python
print(overall_corr)
```

```python
region_corrs
```

```python
region_corrs["Color"] = region_corrs["Correlation"].apply(lambda x: "steelblue" if x <= 0 else "salmon")

# Bar plot of correlations by region
plt.figure(figsize=(10, 6))
sns.barplot(
    data=region_corrs,
    x="Correlation",
    y="Region",
    hue = 'Region',
    palette=region_corrs["Color"].tolist(),
    edgecolor="black"
)

plt.title("Correlation Between Renewables and CO₂ per Capita by Region")
plt.xlabel("Correlation Coefficient")
plt.xlim(-1, 1)
plt.grid(True, axis='x')
plt.tight_layout()
plt.show()
```

```python
# Filter data for Sub-Saharan Africa
ssa_df = merged_df[merged_df["Country Name"] == "Sub-Saharan Africa"]

# Plot time series for Renewables and CO2 per capita
plt.figure(figsize=(14, 5))

# Plot 1: Time series of Renewables and CO2 per capita
plt.subplot(1, 2, 1)
sns.lineplot(data=ssa_df, x="Year", y="Renewables (%)", label="Renewables (%)")
sns.lineplot(data=ssa_df, x="Year", y="CO2 per capita", label="CO₂ per capita")
plt.title("Sub-Saharan Africa: Time Series")
plt.ylabel("Value")
plt.xlabel("Year")
plt.legend()

# Plot 2: Scatter plot of Renewables vs CO2 per capita
plt.subplot(1, 2, 2)
sns.scatterplot(data=ssa_df, x="Renewables (%)", y="CO2 per capita")
plt.title("Sub-Saharan Africa: Renewables vs CO₂ per Capita")
plt.xlabel("Renewables (%)")
plt.ylabel("CO₂ per Capita (tons)")

plt.tight_layout()
plt.show()
```

```python
# Load the GDP per capita dataset
gdp_df = pd.read_csv("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_26433.csv", skiprows=4)

# Filter for the regions of interest
regions_of_interest = [
    "European Union",
    "North America",
    "Middle East & North Africa",
    "Latin America & Caribbean",
    "Sub-Saharan Africa",
    "East Asia & Pacific"
]

gdp_filtered = gdp_df[gdp_df["Country Name"].isin(regions_of_interest)].reset_index(drop=True)

# Drop unnecessary columns
columns_to_drop = ["Country Code", "Indicator Name", "Indicator Code", "Unnamed: 68"]
gdp_filtered = gdp_filtered.drop(columns=[col for col in columns_to_drop if col in gdp_filtered.columns])

# Melt to long format
gdp_long = gdp_filtered.melt(id_vars="Country Name", var_name="Year", value_name="GDP per Capita (US$)")
gdp_long["Year"] = pd.to_numeric(gdp_long["Year"], errors="coerce")
gdp_long = gdp_long.dropna(subset=["Year"])
gdp_long["Year"] = gdp_long["Year"].astype(int)

# Merge with the existing merged_df
merged_with_gdp = pd.merge(merged_df, gdp_long, on=["Country Name", "Year"], how="left")
```
